```python
df=pd.read_csv("/home/ash/Downloads/git/DSBDA/AcademicPerformance.csv",na_values={"na","n/a"})
```

```python
df=df.dropna()
```

```python
df['writing score']=df['writing score'].astype('float')
```

```python
df.dtypes
```

```python
df.isna().sum()
```

```python
df['Total Score']=df[['writing score','reading score','math score']].sum(axis=1)
```

```python
df.describe()
```

```python
q1=df['Total Score'].quantile(0.25)
q3=df['Total Score'].quantile(0.75)
iqr=q3-q1
lower_bound=q1-1.5*iqr
upper_bound=q3+1.5*iqr
```

```python

df=df[(df['Total Score']>=lower_bound)&(df['Total Score']<=upper_bound)]
```
